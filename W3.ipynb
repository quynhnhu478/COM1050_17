Test với: 2000, 1900, 2024, 2023

```python
year = int(input("Nhập năm: "))
# Viết code ở đây
if (year%400==0):
    print(f"Năm {year} là năm nhuận")
elif (year%100==0):
    print(f"Năm {year} là năm không nhuận")
elif (year%4==0):
    print(f"Năm {year} là năm nhuận")
else:
    print(f"Năm {year} là năm không nhuận")
```

**Bài 2:** iết chương trình máy tính:
- Nhập 2 số a, b
- Nhập phép toán (+, -, *, /)
- In kết quả tương ứng
- Với phép chia, kiểm tra b != 0
- Nếu phép toán không hợp lệ, in "Phép toán không hợp lệ"

Ví dụ:
Nhập số thứ nhất: 10
Nhập số thứ hai: 5
Nhập phép toán (+, -, *, /): +
Kết quả: 10 + 5 = 15

```python
a = float(input("Nhập số thứ nhất: "))
b = float(input("Nhập số thứ hai: "))
operator = input("Nhập phép toán (+, -, *, /): ")
# Viết code ở đây
if operator == "+": 
    print(f"{a} {operator} {b} = ", a+b)
elif operator == "-":
    print(f"{a} {operator} {b} = ", a-b)
elif operator == "*":
    print(f"{a} {operator} {b} = ", a*b)
else:
    if b!=0:
        print(f"{a} {operator} {b} = ", a/b)
    else: 
        print("Phép toán không hợp lệ")
```

**Bài 8:**  Viết chương trình tính lương:

Input:
- Số giờ làm việc
- Lương theo giờ
- Loại ngày (weekday/weekend/holiday)
- Đánh giá (poor/average/good/excellent)

Quy tắc:
1. Tính lương theo loại ngày:
   - Weekday:
     * 8 giờ đầu: x1.0
     * Giờ 9-10: x1.5
     * Từ giờ 11: x2.0
   - Weekend: toàn bộ x1.5
   - Holiday: toàn bộ x2.0

2. Thưởng/phạt hiệu suất:
   - poor: -10%
   - average: 0%
   - good: +10%
   - excellent: +20%

3. Thưởng chuyên cần:
   - Làm >= 48 giờ: +500,000đ

Output chi tiết:
- Lương giờ hành chính: ...
- Lương tăng ca: ...
- Tổng lương cơ bản: ...
- Điều chỉnh hiệu suất: ...
- Thưởng chuyên cần: ...
- Tổng lương: ...

```python

hours = float(input("Số giờ làm việc: "))
hourly_rate = float(input("Lương theo giờ: "))
day_type = input("Loại ngày (weekday/weekend/holiday): ")
performance = input("Đánh giá (poor/average/good/excellent): ")
# Viết code ở đây
if day_type=="weekday":
    if hours<=8:
        hanhchinh=hours*hourly_rate*1.0
        tangca=0
    elif hours<=10:
        hanhchinh=8.0*hourly_rate
        tangca=(hours-8)*hourly_rate*1.5
    else:
        hanhchinh=8.0*hourly_rate
        tangca=2*hourly_rate*1.5+(hours-10)*hourly_rate*2.0
elif day_type=="weekend":
    hanhchinh=0
    tangca=hours*hourly_rate*1.5
else: 
    hanhchinh=0
    tangca=hours*hourly_rate*2.0
tong=hanhchinh+tangca
print("\n========== BẢNG TÍNH LƯƠNG ==========")
print(f"Lương giờ hành chính: {hanhchinh:,.0f}(đ)")
print(f"Lương tăng ca: {tangca:,.0f}(đ)")
print(f"Tổng lương cơ bản: {tong:,.0f}(đ)")
if performance=="poor":
    if hours>=48:
        print(f"Điều chỉnh hiệu suất: giảm {tong*0.1:,.0f}(đ)")  
        print("Thưởng chuyên cần: 500,000(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*0.9+500000:,.0f}(đ)")
    else:
        print(f"Điều chỉnh hiệu suất: giảm {tong*0.1:,.0f}(đ)")  
        print("Thưởng chuyên cần: 0(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*0.9:,.0f}(đ)")
    
elif performance=="good":
    if hours>=48:
        print(f"Điều chỉnh hiệu suất: tăng {tong*0.1:,.0f}(đ)")  
        print("Thưởng chuyên cần: 500,000(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*1.1+500000:,.0f}(đ)")
    else:
        print(f"Điều chỉnh hiệu suất: tăng {tong*0.1:,.0f}(đ)")  
        print("Thưởng chuyên cần: 0(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*1.1:,.0f}(đ)")
elif performance=="excellent":
    if hours>=48:
        print(f"Điều chỉnh hiệu suất: tăng {tong*0.2:,.0f}(đ)")  
        print("Thưởng chuyên cần: 500,000(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*1.2+500000:,.0f}(đ)")
    else:
        print(f"Điều chỉnh hiệu suất: tăng {tong*0.2:,.0f}(đ)")  
        print("Thưởng chuyên cần: 0(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong*1.2:,.0f}(đ)")
else: 
    if hours>=48:
        print(f"Điều chỉnh hiệu suất:0(đ)")  
        print("Thưởng chuyên cần: 500,000(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong+500000:,.0f}(đ)")
    else:
        print(f"Điều chỉnh hiệu suất: 0(đ)")  
        print("Thưởng chuyên cần: 0(đ)")
        print("-------------------------------------")
        print(f"Tổng lương: {tong:,.0f}(đ)")

```

Bài 9: Chương trình sau có lỗi. Hãy:
1. Thêm các dòng print để trace control flow
2. Tìm và sửa lỗi
3. Giải thích lỗi là gì?

```python
price = float(input("Giá sản phẩm: "))
quantity = int(input("Số lượng: "))
customer_type = input("Loại khách (VIP/Regular): ")

discount = 0
#trace 1
if customer_type == "VIP":
    print("Khách hàng này thuộc hạng VIP")
    if quantity > 10:
        print("Số lượng khách hàng trên 10 khách")
        discount = 0.2
        print("Giảm giá: 20%") 
    else:
        print("Số lượng khách hàng từ 10 khách trở xuống")
        discount = 0.1
        print("Giảm giá: 10%") 
elif customer_type == "Regular":
    print("Khách hàng này thuộc hạng Regular")
    if quantity > 20:
        print("Số lượng khách hàng trên 20 khách")
        discount = 0.15
        print("Giảm giá 15%")
    else:
        print("Số lượng khách hàng từ 20 khách trở xuống ")
        discount = 0.05
        print("Giảm giá 5%")
#trace 2
print(f"Giá lúc đầu: {price*quantity:,.0f}(đ)")
# Lỗi ở đây!
final_price = (price*quantity)*(1-discount) 
#lỗi tính giá cuối, giá cuối phải bằng tổng giá gốc-tổng giá gốc*giảm giá và rút gọn lại thành như trên 
print(f"Giá cuối: {final_price:,.0f}(đ)") #có thể phân cách các hàng bằng đấu ","để dễ nhìn hơn 
```
