**W5A3**```so nguyen to```

````python
import math
def snt(x):
    if x<2: return False
    for i in range(2,int(math.sqrt(x))+1):
        if x%i==0: return False
    return True 
n=int(input())
if snt(n)==True:
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không là số nguyên tố")
````
