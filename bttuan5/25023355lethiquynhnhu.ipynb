**Bài 1:** Viết hàm triangle_type(a, b, c) nhận vào 3 cạnh và trả về loại tam giác:



*   "Không phải tam giác" nếu không thỏa điều kiện tam giác
*  "Tam giác đều" nếu cả 3 cạnh bằng nhau
*  "Tam giác cân" nếu có 2 cạnh bằng nhau
*   "Tam giác vuông" nếu thỏa định lý Pythagoras
*   "Tam giác thường" trong các trường hợp còn lại






Yêu cầu:

Sử dụng nested if statements
Viết test cases cho mỗi loại tam giác

```python
def triangle_type(a, b, c):
    # Code của bạn ở đây
    if a+b>c and b+c>a and c+a>b:
     if a==b==c: return"Tam giác đều"
     elif a==b or b==c or c==a: return "Tam giác cân"
     elif (a**2+b**2==c**2) or (a**2+c**2==b**2) or(c**2+b**2==a**2): return "Tam giác vuông"
     else: return "Tam giác thường"
    else: return "Không phải tam giác"
a,b,c=map(int,input().split())
print(triangle_type(a,b,c))
```

**Bài 2:** Viết hàm calculate_shipping(weight, distance, is_express) tính phí vận chuyển:
Quy tắc:

Phí cơ bản:



*   Nếu weight <= 1kg: 20,000 đồng
*   Nếu 1kg < weight <= 5kg: 40,000 đồng

*   Nếu weight > 5kg: 60,000 đồng

Phụ phí khoảng cách:

*   Nếu distance > 100km: thêm 10,000 đồng
*   Nếu distance > 500km: thêm 30,000 đồng


Phụ phí express:

Nếu is_express = True: nhân hệ số 1.5



Yêu cầu:

Viết hàm trả về tổng phí vận chuyển
Viết ít nhất 8 test cases để cover các trường hợp

```python
def calculate_shipping(weight, distance, is_express):
    # Code của bạn ở đây
    if weight <=1: s=20000
    elif weight <=5: s=40000
    else: s=60000
    if distance > 500: s+=30000
    elif distance > 100: s+=10000
    if is_express=="True": s*=1.5
    return f"Tổng phí vận chuyển: {s:,.0f}đ"
weight=float(input())
distance=float(input())
is_express=input()
print(calculate_shipping(weight,distance,is_express))
```

**Bài 4:** Viết hàm suggest_outfit(temperature, is_raining, is_windy, event_type) gợi ý trang phục.
Input:


*   temperature: nhiệt độ (°C)

*   is_raining: True/False
*   is_windy: True/False


*   event_type: "formal", "casual", "sport"


Quy tắc:

Xét event_type trước.
Với mỗi loại event, xét thời tiết:


*   Nếu temperature < 15: đề xuất mặc ấm

*   Nếu is_raining: đề xuất áo mưa/ô
*   Nếu is_windy: đề xuất áo khoác




Yêu cầu:

Sử dụng nested if statements
Trả về string mô tả trang phục
Viết 12 test cases để cover tất cả các branch

```python
def suggest_outfit(temperature, is_raining, is_windy, event_type):
    # Code của bạn ở đây
    suggestions = []
    if event_type == "formal":
        print("Mặc đồ trang trọng")
    elif event_type == "casual":
        print("Mặc đồ bình thường")
    elif event_type == "sport":
        print("Mặc đồ thể thao")
    else:
        print("Loại sự kiện không xác định")
    if temperature < 15:
       print("Mặc áo ấm")
    if is_raining=="True":
       print("Mặc áo mưa hoặc đem theo ô")
    if is_windy=="True":
       print("Mặc áo khoác")
    return "\n".join(suggestions)
temperature=float(input())
is_raining=input()
is_windy=input()
event_type=input()
print(suggest_outfit(temperature,is_raining,is_windy,event_type))
```
