**W5A3**```so nguyen to```

````python
import math
def snt(x):
    if x<2: return False
    for i in range(2,int(math.sqrt(x))+1):
        if x%i==0: return False
    return True 
n=int(input())
if snt(n)==True:
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không là số nguyên tố")
````

**W5A5**```tim vi tri bang k dau tien trong day ```

````python
def timk(a,k):
    for i in range(len(a)):
        if a[i]==k:
            return i+1
    return -1
n=list(map(int,input().split()))
k=int(input())
print(timk(n,k))
````

````python
def giaithua(n):
    s=1
    for i in range(2,n+1):
        s*=i
    return s
n=int(input())
print(giaithua(n))
````

**W5A7** ```phep toan```

````python
num1 = float(input("Nhập số thứ nhất: "))
num2 = float(input("Nhập số thứ hai: "))
operat = input("Nhập toán tử (+, -, *, /): ")

if operat == '+':
    s = num1 + num2
elif operat == '-':
    s = num1 - num2
elif operat == '*':
    s = num1 * num2
elif operat == '/':
    if num2 == 0:
        print("Lỗi: không thể chia cho 0")
        exit()
    s = num1 / num2
print(f"Kết quả: {s:.2f}")
````
