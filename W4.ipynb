**W4A8** ```tim so doi xung```

````python
def reverse(n): #hàm tìm số đảo ngược
    s=0
    while(n>0):
        s=s*10+n%10
        n=n//10
    return(s)
n=int(input())
dem=0
while (n>0):#kiểm tra số đối xứng 
    n=n+reverse(n)
    if(n==reverse(n)):
        dem+=1
        break
    else: 
        dem+=1
        continue
print(f"số bước: {dem}")
print(f"số palindrome thu được là: {reverse(n)}")
````

**W4A10** ```day collatz```
###### nếu n chẵn, số tiếp theo là n//2, nếu n lẻ số tiếp theo là 3n+1 và tính đến khi nào chạy đến 1 
###### in ra số nhỏ nhất có dãy collatz dài nhất trong đoạn [1,n] và độ dài của nó 


````python
import math
def collatz(x):
     dem=1
     while x>1:
          if x%2==0:
               x=x//2
               dem+=1
          else:
               x=3*x+1
               dem+=1
     return (dem)
n=int(input())
maxx=0
minn=0
for i in range(1,n+1):
     if(collatz(i)>=maxx):
          maxx=collatz(i)
          minn=i
     if(collatz(i)==maxx and i<minn):
          minn=i
print(minn,maxx)
````

**W4A13** ```cặp số thân thiết```
- tổng các ước trừ chính nó của số a = số b và ngược lại 
    ###### ví dụ: 220 và 284, 220 có các ước 1 2 4 5 10 11 20 44 55 110, tổng = 284 và ngược lại 
    

````python
import math
def tonguoc(x):
     s=0
     for i in range(1,int(math.sqrt(x))):
          if x%i==0:
               s+=i
               j=x//i
               if j!=i and j!=x: 
                    s+=j
     return(s)
a,b=map(int,input().split())
if tonguoc(a)==b and tonguoc(b)==a:
     print("true")
else:
     print("false")
````
